```python
import re

def clean_text(text):
    text = text.lower()  # Chuyển tất cả ký tự thành chữ thường
    text = re.sub(r'\d+', '', text)  # Loại bỏ các chữ số
    text = re.sub(r'[^\w\s]', '', text)  # Loại bỏ dấu câu
    return text

# Áp dụng hàm làm sạch cho cột văn bản
df['Content'] = df['Content'].apply(clean_text)

from underthesea import word_tokenize

# Tách từ cho cột văn bản
df['tokenized_content'] = df['Content'].apply(lambda x: word_tokenize(x))
df.head()
```
